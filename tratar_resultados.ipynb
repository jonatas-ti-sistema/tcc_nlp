```python
def calcular_significancia(perfil_a, perfil_b):
    acc_a = df[df['Nome_Perfil'] == perfil_a]['Acuracia']
    acc_b = df[df['Nome_Perfil'] == perfil_b]['Acuracia']
    
    t_stat, p_val = stats.ttest_ind(acc_a, acc_b)
    
    print(f"Comparando {perfil_a} vs {perfil_b}:")
    print(f"P-valor: {p_val:.4f}")
    if p_val < 0.05:
        print("✅ Resultado Estatisticamente Significante (p < 0.05)")
    else:
        print("❌ Não há evidências estatísticas de diferença.")

# Exemplo de uso:
calcular_significancia('Perfil_1', 'Perfil_4')
```

```python
from scipy import stats

# 1. Filtrar os dados de acurácia (binária: 0 ou 1) para cada perfil
grupo_perfil2 = df_u[df_u['Nome_Perfil'] == 'Perfil 2']['Acuracia']
grupo_perfil4 = df_u[df_u['Nome_Perfil'] == 'Perfil 4']['Acuracia']

# 2. Executar o Teste T de Student (Independente)
t_stat, p_valor = stats.ttest_ind(grupo_perfil2, grupo_perfil4)

print(f"Estatística T: {t_stat:.4f}")
print(f"p-valor: {p_valor:.4f}")

# 3. Interpretação automática
alfa = 0.05
if p_valor < alfa:
    print("Resultado: Estatisticamente significante (Há diferença real)")
else:
    print("Resultado: Não há diferença estatística significante (Diferença por acaso)")
```

```python
def calcular_significancia(perfil_a, perfil_b):
    acc_a = df_u[df_u['Nome_Perfil'] == perfil_a]['Acuracia']
    acc_b = df_u[df_u['Nome_Perfil'] == perfil_b]['Acuracia']
    
    t_stat, p_val = stats.ttest_ind(acc_a, acc_b)
    
    print(f"Comparando {perfil_a} vs {perfil_b}:")
    print(f"P-valor: {p_val:.4f}")
    if p_val < 0.05:
        print("✅ Resultado Estatisticamente Significante (p < 0.05)")
    else:
        print("❌ Não há evidências estatísticas de diferença.")

# Exemplo de uso:
calcular_significancia('Perfil_2', 'Perfil_4')
```
